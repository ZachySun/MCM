```python
import pandas as pd
import numpy as np
from numpy import nan
```

```python
# 生成含缺失值的数据
data=pd.DataFrame(np.arange(3,35,1).reshape(8,4),columns=list('abcd'))
data.iloc[1:2,:3]=nan
data
```

    1.检查缺失值

```python
data.isnull()
```

```python
data.notnull()
```

```python
data.isnull().sum()
```

```python
data.info()
```

    2.缺失值填充

```python
data_1 = data.fillna(0)
data_1
```

```python
data_2 = data.fillna("$")
data_2
```

```python
data['a']=data['a'].fillna(data['a'].sum())
data
```

```python
# 插值算法
df = pd.DataFrame([(0.0, np.nan, -1.0, 1.0),
                   (np.nan, 3.0, np.nan, np.nan),
                   (2.0, 3.0, np.nan, 9.0),
                   (np.nan, 4.0, -4.0, 15.0)],
                  columns=list('abcd'))
df
```

```python
df.interpolate(method='linear', limit_direction='forward', axis=0)
```

```python
df.interpolate(method='linear', axis=0)
```

```python
df['d'].interpolate(method='spline', axis=0, order=2)
```
